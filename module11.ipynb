```python
# lambda 

lambda functions are a key concept in Python and extremely useful for short, 
throwaway functions — especially in sorting, filtering, and mapping data.

lambda arguments: expression
lambda — keyword

arguments — inputs (like parameters)

expression — single expression whose result is returned automatically

def sqr(x):
    return x**2
sqr(5)    

sqr1=lambda x:x**2
sqr1(5)


square = lambda x: x**2
print(square(5))   # Output: 25

def square(x):
    return x**2
square(5)    



#Lambda with Multiple Arguments
add = lambda a, b: a + b
print(add(3, 7))  # Output: 10

type(add)

#  Where Lambdas Are Commonly Used

1. ✅ In sorted() or .sort()
✅ With map()
3. ✅ With filter()

# no return value
#type(a) give s function
#one line code
#not use for code reusability
#no name

#Limitations of Lambda
Only one expression (no statements, loops, or multiple lines)
Often best used for simple operations
For complex logic → use a regular def function

Sorting by Custom Rule (using key)

l=['apple','bannan','cherry']
b=lambda l:l[0]
print(b(l))

def log(x):
    return x[1]=="a"
log("apple")    

b=lambda x:x[0]=="a"
print(b("apple"))
print(b("banana"))

b=lambda x:"even" if x%2==0 else "odd"
print(b(2))
print(b(3))

#lambda with map
print(l)
a=list(map(lambda x:x+2,l))
print(a)

a=list(map(lambda x:x%2==0,l))
print(a)

people = [
    {"name": "Alice", "age": 25},
    {"name": "Bob", "age": 30},
    {"name": "Charlie", "age": 20}
]
list(map(lambda x:x["name"],people))

l1=[1,2,3,4,5]
l2=[6,7,8,9,4]
list(map(lambda x,y:x+y,l1,l2))

#lambda with filter
print(l)
list(filter(lambda x:x%2==0,l)) # gives only true part

# lambda with reduce
#it reduce the list
from functools import reduce

l = [1, 2, 3, 4]
result = reduce(lambda x, y: x + y, l)
print(result)   # 10


result=reduce(lambda x,y:x if x>y else y,l)
print(result)

#1,2--->2,3---->3,4----->4

word=["hello"," ","world","!"]
reduce(lambda x,y:x+y,word)



# practice some question

words = ["apple", "Banana", "cherry"]
sorted(words)

words = ["apple", "Banana", "cherry"]
sorted_words = sorted(words, key=lambda x: x.lower())
print(sorted_words)  # ['apple', 'Banana', 'cherry']

# sorting a list of tuples —

#Sort by the First Element
data = [(3, 'apple'), (1, 'banana'), (2, 'cherry')]
sorted_data = sorted(data)
print(sorted_data)
#first by the first element, then the second if the first is equal.

#Sort by the Second Element
#Use the key argument to tell Python what to sort by.
#lambda used As a key for sorting: 


data = [(3, 'apple'), (1, 'banana'), (2, 'cherry')]
sorted_data = sorted(data, key=lambda x: x[1])
print(sorted_data)

#Sort by the Second Element (Descending)
data = [(3, 'apple'), (1, 'banana'), (2, 'cherry')]
sorted_data = sorted(data, key=lambda x: x[1], reverse=True)
print(sorted_data)

#Sort by Length of the Second Element
data = [(3, 'apple'), (1, 'kiwi'), (2, 'banana')]
sorted_data = sorted(data, key=lambda x: len(x[1]))
print(sorted_data)

#Sort by Multiple Keys
#You can sort by more than one thing — for example, first by number, then by name:

data = [(2, 'banana'), (2, 'apple'), (1, 'cherry')]
sorted_data = sorted(data, key=lambda x: (x[0], x[1]))
print(sorted_data)

#Sort List of Tuples Containing Mixed Types
#If you have tuples like (int, str) and want to sort safely even if elements differ in type:

data = [(3, 'apple'), ('1', 'banana'), (2, 'cherry')]
sorted_data = sorted(data, key=lambda x: (str(x[0]), x[1]))
print(sorted_data)

# sorting lists of dictionaries or nested lists

#Sorting a List of Dictionaries
#Sort by a Single Key
people = [
    {"name": "Alice", "age": 25},
    {"name": "Bob", "age": 30},
    {"name": "Charlie", "age": 20}
]

sorted_people = sorted(people, key=lambda x: x["age"])  #{"name": "Alice", "age": 25},  25,30,20
print(sorted_people)

#Sort by a String Key (Alphabetically)
sorted_people = sorted(people, key=lambda x: x["name"])
print(sorted_people)

#Sort by Multiple Keys
#You can sort by one key primarily, then another key for tie-breaking.

people = [
    {"name": "Alice", "age": 25},
    {"name": "Alice", "age": 22},
    {"name": "Bob", "age": 30}
]

sorted_people = sorted(people, key=lambda x: (x["name"], x["age"]))
print(sorted_people)

#Sort in Reverse Order
sorted_people = sorted(people, key=lambda x: x["age"], reverse=True)

# Sorting a Nested List

#Sort by First Element
data = [[3, 'apple'], [1, 'banana'], [2, 'cherry']]
sorted_data = sorted(data, key=lambda x: x[0])
print(sorted_data)

#Sort by Second Element
sorted_data = sorted(data, key=lambda x: x[1])
print(sorted_data)

#Sort by Computed Value
#You can use any function in the key.
#Example — sort by length of the second element:

data = [[1, 'kiwi'], [2, 'banana'], [3, 'apple']]
sorted_data = sorted(data, key=lambda x: len(x[1]))
print(sorted_data)
```
