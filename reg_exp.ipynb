```python
What Is a Regular Expression (Regex)?

A Regex is a mini-language used to search, match, and extract patterns inside text.

It helps you find things like:

Only letters

Only numbers

Email patterns

Dates

Words starting with “A”

Anything that matches a rule
```

```python
[A-Za-z]	   Any letter	              A, d, Z
[0-9]	        Any digit	                    3
[A-Za-z0-9]  	Letters or digits	      A, 9
[^A-Za-z]	    NOT a letter	         1, -, @
```

```python
df["col"].str.replace(r"[^A-Za-z]", "", regex=True)
```

```python
\d	one digit (0–9)
\d+	one or more digits
\d{4}   exactly 4 digit
```

```python
# Special Characters
.	Any character	a, 5, @
\d	Digit (0–9)	     1
\w	Word character (letters, digits, _)     	A, 5
\s	Whitespace	                            space, tab
\D	Not a digit	
\W	Not a word char
```

```python
^	Start of string
$	End of string
{n}
Exactly n
```

```python
Match email only if the entire string is valid:
r"^[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Za-z]{2,}$"
```

```python
match string starting with “AB”:

^AB


Example: match string ending with digits:

\d+$
```

```python
Email extraction example

Extract username before @:

([^@]+)@
Groups & Extracting
(a) Parentheses capture part of the string
df["email_user"] = df["email"].str.extract(r"([^@]+)@")
([^@]+) → capture everything until @
@ → match the @ symbol
```

```python
Check if valid email:

^[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Za-z]{2,}$
```

```python
Extract letters from code
df["code_letters"] = df["code"].str.replace(r"[^A-Za-z]", "", regex=True)

Extract email user
df["email_user"] = df["email"].str.extract(r"([^@]+)@")

Check valid email
df["valid_email"] = df["email"].str.contains(
    r"^[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Za-z]{2,}$"
)
```
