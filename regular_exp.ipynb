```python
We use regex to:

👉 Search
👉 Validate Email validation
👉 Extract  Extract Data from Text
👉 Replace
👉 Clean text
```

```python
from re import re
```

```python
Python does NOT support regex directly by default.
So we must import the module to use regex functions.
```

```python
text = "Python I am learning Python Python "

result = re.findall("(Python)+",text)
print(result)
if result:
    print("Found!")
else:
    print("Not Found")
```

```python
Function	Purpose
re.search()	Finds first match anywhere in string
re.match()	Checks match at beginning of string
re.findall()	Returns all matches as a list
re.sub()	Replace matched text
re.split()	Split string using pattern
```

```python
Symbol	Meaning	Example
.	Any character except newline	a.c → abc, axc
^	Start of string	^Hello
$	End of string	world$
*	0 or more times	ab*  a,ab,abb
+	1 or more times	ab+  ab,abb
?	0 or 1 time	ab?   previous letter
\d	Digit (0–9)	\d\d  [0-9]  [a z 0-9
\w	Word character	letters, digits
\s	Space	
[]	Set of characters	[abc]
{4}fixed
```

```python
# extract only number
```

```python
text = "My marks are 85 and 90"

numbers = re.findall("\d+", text)

print(numbers)
```

```python
\d+ means:

\d → digit

+ → one or more
```

```python
# valid email
```

```python
email = "renu89@gmail.com"

pattern = r"^\w+@\w+\.\w+$"

if re.match(pattern, email):
    print("Valid Email")
else:
    print("Invalid Email")   # beginning only
```

```python
text = "My email is renu@gmail.com"

pattern = r"\w+@\w+\.\w+"

result = re.search(pattern, text)
print(result)   # anywhere
```

```python
print(result.group())
```

```python
# valid phone number
```

```python
text = "My number is 89999999999876543210"

result = re.findall(r"\b\d{10}\b", text)
print(result)
```

```python
Has exactly 10 digits
re.findall():

Returns ALL matches
Is not attached to other numbers or characters
```

```python
# Validate Indian PAN Card
```

```python
Format: 5 letters + 4 digits + 1 letter
Example: ABCDE1234F

pan = "ABCDE1234F"

pattern = r"^[A-Z]{5}[0-9]{4}[A-Z]$"

if re.match(pattern, pan):
    print("Valid PAN")
else:
    print("Invalid PAN")
```

```python
import pandas as pd

data = {
    "name": ["     Renu", "Amit", "Priya", "John123"],
    "email": ["renu@gmail.com", "amit@yahoo.in", "invalid_email", "john123@gmail.com"],
    "phone": ["9876543210", "12345", "9999999999", "abc9876543"]
}

df = pd.DataFrame(data)
df
```

```python
df["numbers_in_name"] = df["name"].str.extract(r"(\d+)")

df
```
