```python
def decorator(func):
    def wrapper(a,b):
        print("Start of function")
        result = func(a,b)  # call original function and save result
        print("End of transaction")
        return result  # return the result to the caller
    return wrapper

# Example function
```

```python
def add(a, b):
    return a + b
```

```python
# Call the decorated function
decorated_add=decorator(add(5,3))
```

```python
decorator(add)(5,3)
```

```python
decorated_add(5, 3)
#print("Result:", result)
```

```python
Why wrapper() needs *args, **kwargs
```

```python
def my_decorator(func):
    def wrapper(*args, **kwargs):
        print("Before")
        func(*args, **kwargs)
        print("After")
    return wrapper
@my_decorator
def add(a, b):
    print(a + b)

add(2, 3)
```
